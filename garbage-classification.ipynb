```python
# Importing Modules 
import pandas as pd
import numpy as np
import cv2
import os
from pathlib import Path
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
directory = Path("../input/garbage-classification/Garbage classification/Garbage classification")
files = list(directory.glob(r"**/*.jpg"))
files
```

```python
# Getting labels from files
labels = list(map(lambda x: os.path.split(os.path.split(x)[0])[1],files))
labels
```

```python
# Loading Data
image_data = pd.DataFrame({"FilePath":files,"Labels":labels}).astype(str) 
```

```python
image_data
```

```python
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.models import Model 
from keras import layers
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_data,test_data = train_test_split(image_data,test_size = 0.3,random_state = 42,shuffle = True)
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,validation_split = 0.2)
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
train_images = train_datagen.flow_from_dataframe(dataframe = train_data,target_size = (224,224),x_col = "FilePath",y_col = "Labels",class_mode = "categorical",subset = "training")
val_images = train_datagen.flow_from_dataframe(dataframe = train_data,traget_size = (224,224),x_col = "FilePath",y_col = "Labels",class_mode = "categorical",subset = "validation")
test_images = test_datagen.flow_from_dataframe(dataframe = test_data,targte_size = (224,224),x_col = "FilePath",y_col = "Labels",class_mode = "categorical")
```

```python
# Building Neural Network
model = keras.Sequential([
    keras.layers.Conv2D(input_shape = (224,224,3),filters = 32,kernel_size = (3,3),padding = "same",activation = "relu"),
    keras.layers.MaxPool2D(pool_size = (2,2),strides = (2,2)),
    keras.layers.Conv2D(filters = 64,kernel_size = (3,3),activation = "relu",padding = "same"),
    keras.layers.MaxPool2D(pool_size = (2,2),strides = (2,2)),
    keras.layers.GlobalAveragePooling2D(),
    keras.layers.Dense(128,activation = "relu"),
    keras.layers.Dense(128,activation = "relu"),
    keras.layers.Dense(6,activation = "softmax")
])
```

```python
# Compiling the Model
model.compile(optimizer = "adam",metrics = ["accuracy"],loss = "categorical_crossentropy")
```

```python
# Fitting the model
history = model.fit(train_images,validation_data = val_images,epochs = 10)
```

```python
from keras.applications import InceptionV3
```

```python
inception = InceptionV3(include_top=False, weights='imagenet', input_shape=(100,100,3))
inception.summary()
```

```python
for layer in inception.layers:
    layer.trainable=False
```

```python
x = keras.layers.GlobalAveragePooling2D()(inception.output)
x = keras.layers.Dense(512,activation='relu')(x)
x = keras.layers.Dense(512,activation='relu')(x)
prediction = keras.layers.Dense(6,activation='softmax')(x)
model = Model(inputs=inception.input, outputs=prediction)
model.summary()
```

```python
model.compile(optimizer = "adam",metrics = ["accuracy"],loss = "categorical_crossentropy")
```

```python
# Fitting the model
history = model.fit(train_images,validation_data = val_images,epochs = 10)
```

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("Accuracy_Graphs")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend(["Accuracy","Val_Accuracy"])
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Loss_Graphs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend(["loss","Val_loss"])
```
